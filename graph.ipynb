```python
import matplotlib.pyplot as plt

# Data
committed = [8]
completed = [8]

# Bar configuration
bar_width = 0.3
gap_between_bars = 0.4

# Bar positions
committed_pos = [0]
completed_pos = [committed_pos[0] + bar_width + gap_between_bars]

# Combine positions and labels
bar_positions = committed_pos + completed_pos
bar_labels = ['Committed', 'Completed']
bar_values = committed + completed
bar_colors = ['skyblue', 'seagreen']

# Plot bars
fig, ax = plt.subplots()
bars = ax.bar(bar_positions, bar_values, bar_width, color=bar_colors)

# Set x-ticks to each bar and label them
ax.set_xticks(bar_positions)
ax.set_xticklabels(bar_labels)

# Labels and title
ax.set_ylabel('Number of User Stories')
ax.set_title('User Stories - Sprint 3')
ax.legend(bars, ['Committed User Stories', 'Completed User Stories'],loc='lower right')

# Display the graph
plt.tight_layout()
plt.show()
```
